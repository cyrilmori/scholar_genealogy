# Test notebook for Scholar Genealogy
A script which uses the Google Scholar API to identify mentor/student relationships (academic genealogy) within a field. Made for personal purposes, specifically in the field of quantum physics.

```python
from bs4 import BeautifulSoup
import feedparser
```

```python
author_name = 'M. H. Devoret'

search_author = 'au:' + '_'.join(author_name.split()).lower()
print(search_author)
```

```python
base_url = 'http://export.arxiv.org/api/query?search_query='
max_results = 1000
rss_request = base_url + search_author + '&max_results=' + str(max_results)
print(rss_request)
```

```python
d = feedparser.parse(rss_request)
print(len(d.entries))
print(d.entries[0].keys())
print(d.entries[0]['published'])
print(d.entries[0]['authors'])
```

```python
for i in range(len(d.entries)):
    entry = d.entries[i]
    authors = entry['authors']
    # if author_name in authors[0]['name']:
    #     print(str(i) + ':\t' + str(authors))
    if author_name in authors[-1]['name']:
        print(str(i) + ':\t' + str(authors))
        
```
